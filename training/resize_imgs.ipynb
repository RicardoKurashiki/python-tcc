```python
import PIL as image_lib
import PIL.Image
import pathlib
import shutil
import os

train_path = "../datasets/libras/train/"
test_path  = "../datasets/libras/test/"

classes = os.listdir(train_path)
classes = sorted(classes)
n_classes = len(classes)
print(f"Classes: {classes}")
file_count = []

for fld in os.listdir(train_path):
    crt = os.path.join(train_path, fld)
    image_count = len(os.listdir(crt))
    file_count.append(image_count)
    print(f'{crt} contains {image_count} images')

print(f'\nTotal number of images: {sum(file_count)}')

# -------------------- #
# TRAIN PATH
# Set the target size (75x75) for the InceptionV3 model
image_size = (64, 64)
target_size = (64, 64)

im_per_class = min(file_count)
input_directory = train_path
output_directory = f'../datasets/libras/{target_size[0]}x{target_size[1]}/train/'

# iterate through all folders (there should be one folder per object class)
for folder in os.listdir(input_directory):
  # create the output folder for processed images for current class
  # delete folder and contents if there is one already
  out = os.path.join(output_directory, folder)
  if os.path.exists(out):
      shutil.rmtree(out)
  os.makedirs(out)


  path = pathlib.Path(os.path.join(input_directory, folder))
  num_images = 0
  for file in list(path.glob('*')):
      # open image, center crop to a square
      # save to the output folder
      with PIL.Image.open(file) as im:
          crop_im = im.resize(target_size)
          crop_im = crop_im.convert('RGB')
          crop_im.save(os.path.join(out, str(num_images) + '.png'))
      # break when desired number of images
      # has been processed (to keep classes balance)
      num_images = num_images + 1
      if (num_images > im_per_class):
        print(f"{path} resized from {image_size} to {target_size}")
        break
      
# TEST PATH
input_directory = test_path
output_directory = f'../datasets/libras/{target_size[0]}x{target_size[1]}/test/'

# iterate through all folders (there should be one folder per object class)
for folder in os.listdir(input_directory):
  # create the output folder for processed images for current class
  # delete folder and contents if there is one already
  out = os.path.join(output_directory, folder)
  if os.path.exists(out):
      shutil.rmtree(out)
  os.makedirs(out)


  path = pathlib.Path(os.path.join(input_directory, folder))
  num_images = 0
  for file in list(path.glob('*')):
      # open image, center crop to a square
      # save to the output folder
      with PIL.Image.open(file) as im:
          crop_im = im.resize(target_size)
          crop_im = crop_im.convert('RGB')
          crop_im.save(os.path.join(out, str(num_images) + '.png'))
      # break when desired number of images
      # has been processed (to keep classes balance)
      num_images = num_images + 1
      if (num_images > im_per_class):
        print(f"{path} resized from {image_size} to {target_size}")
        break
```
